```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
#italy covid Death
df1 = pd.read_csv('total-deaths-covid-19.csv')
df1.Date = pd.to_datetime(df1['Date'] ,format='%b %d, %Y')
df1 = df1[(df1.Code == 'ITA')]
                         
#df1=df1[df1.Date.str.match('(.*Mar.*)').str.len()>0]
#df1=df1['Date'].apply(lambda x: x.replace(',',''))
#df1.describ
start_date = '2020-03-01'
end_date = '2020-03-31'
death = df1[ (df1['Date'] >= start_date) & (df1['Date'] <= end_date)]
death =death.rename(columns ={'Total confirmed deaths due to COVID-19 (deaths)':'Death'})


#italy covid Death
df = pd.read_csv('coronavirus-cfr.csv')
df.Date = pd.to_datetime(df['Date'] ,format='%b %d, %Y')
df = df[(df.Code == 'ITA')]
                         
#df1=df1[df1.Date.str.match('(.*Mar.*)').str.len()>0]
#df1=df1['Date'].apply(lambda x: x.replace(',',''))
#df1.describ
start_date = '2020-03-01'
end_date = '2020-03-31'
Italy_covid_rate = df[ (df['Date'] >= start_date) & (df['Date'] <= end_date)]

rate = Italy_covid_rate.rename(columns ={'Case fatality rate of COVID-19 (%) (Only observations with â‰¥100 cases) (%)':'Rate'})
rate
```

```python
fig, ax1 = plt.subplots()
ax1.plot(death['Date'].dt.day,death['Death'])
ax1.set_xlabel('Day')
ax1.set_ylabel('Cumulative Total Death')
ax1.tick_params('y')

# frameon turns the frame around the legend on. framealpha adjusts the transparency of the box.
ax1.legend(loc='upper left', fontsize=9, frameon=True, framealpha=1)

ax2 = ax1.twinx()
ax2.plot(rate['Date'].dt.day,rate['Rate'], 'r', label='Death Rate')
ax2.set_ylabel('% Death Rate')
ax2.tick_params('y')
ax2.legend(loc='upper right', fontsize=9, frameon=True, framealpha=1)

plt.title('COVID-19 Total Death and Death rate in Italy , March-2020')
fig.tight_layout()
plt.show()
```
